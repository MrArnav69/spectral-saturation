# Experiment 3: Ablation Studies

This notebook reproduces the two ablation studies from the paper:
1. **PCA Dimensionality Ablation**: How does the PCA target dimensionality ($d \in \{20, 50, 100\}$) affect accuracy and effective rank?
2. **Regularization Ablation**: Does regularizing the Logistic Regression classifier ($C \in \{\infty, 1.0, 0.1\}$) alleviate saturation or change the asymptotic accuracy?

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import sys

# Add src to python path to allow importing modules
sys.path.append(os.path.abspath('..'))

from src.experiment import load_results
from src.visualization import plot_pca_ablation
```

## 1. PCA Dimensionality Ablation

We compare training curves (Accuracy and Effective Rank vs. $K$) on the MNIST 3v8 task across three different PCA targets ($d=20$, $d=50$, $d=100$).

```python
pca_results_path = '../results/ablation_pca_results.json'
ablation_pca_results = load_results(pca_results_path)

if ablation_pca_results is None:
    raise FileNotFoundError(f"Run 'run_all.py' script first to populate '{pca_results_path}'")

for d, results in ablation_pca_results.items():
    print(f"PCA d={d}:")
    print(f"  erank_inf = {results[-1]['mean_erank']:.2f}")
    print(f"  peak_acc  = {max([r['mean_acc'] for r in results]):.4f}")
```

## 2. Regularization Ablation

We evaluate classification accuracy at $K=4096$ under different inverse regularization strengths $C$ ($C=\infty$ (no regularization), $C=1.0$, $C=0.1$).

```python
reg_results_path = '../results/ablation_reg_results.json'
ablation_reg_results = load_results(reg_results_path)

if ablation_reg_results is None:
    raise FileNotFoundError(f"Run 'run_all.py' script first to populate '{reg_results_path}'")

for C, res in ablation_reg_results.items():
    print(f"C={C:>5s}: accuracy = {res['mean_acc']:.4f} ± {res['std_acc']:.4f}")
```
